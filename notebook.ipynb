#### Melihat Tren Penyewaan Sepeda

```python
mapping_yr = {0: '2011', 1: '2012'}

mapping_mnth = {1: 'January', 2: 'February', 3: 'March', 4: 'April', 5: 'May', 6: 'June',
                7: 'July', 8: 'August', 9: 'September', 10: 'October', 11: 'November', 12: 'December'}

day_df['yr'] = day_df['yr'].map(mapping_yr)
day_df['mnth'] = day_df['mnth'].map(mapping_mnth)

monthly_trend = day_df.groupby(['yr', 'mnth'])['cnt'].agg(['sum', 'mean']).reset_index()
monthly_trend.rename(columns={'sum': 'Total Rentals', 'mean': 'Avg Rentals per Day'}, inplace=True)

order_months = ["January", "February", "March", "April", "May", "June", 
                "July", "August", "September", "October", "November", "December"]
monthly_trend['mnth'] = pd.Categorical(monthly_trend['mnth'], categories=order_months, ordered=True)

print(monthly_trend)
```

Dapat dilihat, penyewaan sepeda per bulan pada tahun 2011 dan 2012 menunjukkan pola yang cukup jelas, dengan jumlah penyewaan tertinggi terjadi pada musim panas (JuniAgustus) dan terendah pada musim dingin (DesemberFebruari). Pada tahun 2011, puncak penyewaan terjadi pada bulan Juni dengan 143.512 penyewaan, sedangkan jumlah terendah tercatat di bulan Januari dengan 38.189 penyewaan. Tren serupa terlihat di tahun 2012, tetapi dengan peningkatan signifikan dalam jumlah penyewaan, di mana September mencatat angka tertinggi sebesar 218.573 penyewaan. 

```python
day_df.groupby("yr").aggregate(
    dteday_nunique=("dteday", "nunique"),
    cnt_max=("cnt", "max"),
    cnt_min=("cnt", "min"),
    cnt_mean=("cnt", "mean"),
    cnt_std=("cnt", "std")
)
```

Dapat dilihat, tahun 2012 memiliki jumlah penyewaan harian yang lebih tinggi dibandingkan tahun 2011, baik dalam rata-rata maupun nilai maksimum. Selain itu, jumlah penyewaan harian di tahun 2012 lebih bervariasi dibandingkan tahun sebelumnya, yang terlihat dari standar deviasi yang lebih besar. Jumlah hari dalam dataset juga sesuai dengan jumlah hari dalam setahun, yaitu 365 untuk 2011 dan 366 untuk 2012 karena tahun kabisat.

#### Melihat Musim Dengan Penyewaan Tertinggi

```python
day_df.groupby("season").aggregate(
    dteday_nunique=("dteday", "nunique"),
    cnt_max=("cnt", "max"),
    cnt_min=("cnt", "min"),
    cnt_mean=("cnt", "mean"),
    cnt_std=("cnt", "std")
)
```

Musim dengan penyewaan sepeda tertinggi adalah musim gugur (fall) dengan rata-rata sekitar 5.644 pengguna per hari, diikuti oleh musim panas (summer) dengan 4.992 pengguna per hari. Musim dingin (winter) juga memiliki jumlah penyewaan yang cukup tinggi, sekitar 4.728 pengguna per hari, tetapi memiliki nilai minimum yang sangat rendah, menunjukkan bahwa ada hari-hari dengan hampir tidak ada penyewaan. Sementara itu, musim semi (spring) menjadi musim dengan penyewaan terendah, rata-rata hanya 2.604 pengguna per hari.

```python
day_df.groupby("weathersit").aggregate(
    dteday_nunique=("dteday", "nunique"),
    cnt_max=("cnt", "max"),
    cnt_min=("cnt", "min"),
    cnt_mean=("cnt", "mean"),
    cnt_std=("cnt", "std")
)
```

Terlihat saat cuaca baik (kategori 1), jumlah penyewaan sepeda rata-rata lebih tinggi dibandingkan kondisi cuaca yang lebih buruk. Saat cuaca sedang (kategori 2), jumlah penyewaan sedikit lebih rendah, dan pada cuaca buruk (kategori 3), jumlah penyewaan menurun drastis. Selain itu, jumlah hari dengan cuaca baik jauh lebih banyak dibandingkan hari dengan cuaca buruk. Hal ini menunjukkan bahwa kondisi cuaca berpengaruh signifikan terhadap jumlah sepeda yang disewa, di mana orang lebih cenderung bersepeda saat cuaca lebih baik. 

## Visualization & Explanatory Analysis

### Pertanyaan 1:

```python
sns.set_style("whitegrid")

monthly_trend = monthly_trend.sort_values(by=["yr", "mnth"])

plt.figure(figsize=(12, 6))

sns.lineplot(data=monthly_trend, x="mnth", y="Total Rentals", hue="yr", marker="o", linewidth=2)

plt.title("Tren Penyewaan Sepeda per Bulan (2011 vs 2012)", fontsize=14)
plt.xlabel("Bulan", fontsize=12)
plt.ylabel("Total Rentals", fontsize=12)
plt.xticks(rotation=45)  # Rotasi label bulan agar lebih mudah dibaca
plt.legend(title="Tahun")

plt.show()
```

1. Peningkatan Penyewaan di Tahun 2012
* Secara keseluruhan, jumlah penyewaan sepeda di tahun 2012 lebih tinggi dibandingkan tahun 2011 di setiap bulan.
* Hal ini menunjukkan adanya peningkatan permintaan terhadap layanan penyewaan sepeda.
  
2. Pola Musiman dalam Penyewaan
* Penyewaan sepeda cenderung mengalami peningkatan dari bulan Januari hingga puncaknya di bulan Juni hingga September, sebelum mulai menurun kembali pada akhir tahun.
* Tren ini dapat dilihat pada kedua tahun, menunjukkan bahwa musim dan cuaca yang mungkin menjadi faktor utama dalam menentukan jumlah penyewaan.

3. Perbedaan Pola Antara 2011 dan 2012
* Di tahun 2011, peningkatan penyewaan terjadi secara bertahap dari Januari hingga Juni, lalu stagnan dari Juli hingga September, sebelum akhirnya menurun.
* Sebaliknya, di tahun 2012, peningkatan penyewaan terjadi dengan lebih cepat, terutama dari Februari ke Maret, dengan lonjakan signifikan pada bulan April.
* Puncak penyewaan di 2012 juga lebih tinggi dan bertahan lebih lama dibandingkan 2011.
  
4. Penurunan Penyewaan pada Akhir Tahun
* Pada kedua tahun, jumlah penyewaan mengalami penurunan yang cukup tajam setelah bulan September, dengan titik terendah di bulan Januari.

### Pertanyaan 2:

```python

# Mengelompokkan dan menjumlahkan rental berdasarkan musim
seasonal_counts = day_df.groupby('season')['cnt'].sum().sort_values(ascending=False)

# Memetakan kode musim ke nama musim
season_mapping = {1: 'Spring', 2: 'Summer', 3: 'Fall', 4: 'Winter'}
seasonal_counts.index = seasonal_counts.index.map(season_mapping)

# Warna untuk setiap musim
season_colors = {
    'Spring': 'lightgreen',
    'Summer': 'gold',
    'Fall': 'lightcoral',
    'Winter': 'lightskyblue'
}

# Membuat plot
plt.figure(figsize=(10, 6))
sizes = seasonal_counts.values
labels = seasonal_counts.index
colors = [season_colors[label] for label in labels]

# Membuat bar chart dengan edgecolor hitam
bars = plt.bar(labels, sizes, color=colors, edgecolor='black')

# Menambahkan angka di atas setiap batang dengan format angka lebih rapi
for bar in bars:
    height = bar.get_height()
    plt.text(bar.get_x() + bar.get_width() / 2, height + 10000, 
             f"{height:,}", ha='center', va='bottom', fontsize=12, fontweight='bold')

# Menambahkan label dan judul
plt.xlabel('Season', fontsize=12)
plt.ylabel('Total Rentals', fontsize=12)
plt.title('Total Bike Rentals by Season', fontsize=14)

plt.tight_layout()
plt.show()
```

Dapat dilihat, jumlah penyewaan sepeda bervariasi berdasarkan musim. Musim gugur (Fall) memiliki jumlah penyewaan tertinggi, yaitu 1.061.129, kemungkinan karena cuaca masih nyaman sebelum memasuki musim dingin. Musim panas (Summer) berada di posisi kedua dengan 918.589 penyewaan, yang dapat disebabkan oleh cuaca cerah dan meningkatnya aktivitas luar ruangan. Musim dingin (Winter) menempati posisi ketiga dengan 841.613 penyewaan, menunjukkan bahwa meskipun suhu lebih dingin, masih banyak orang yang menggunakan sepeda sebagai transportasi utama. Sementara itu, musim semi (Spring) memiliki jumlah penyewaan terendah, yaitu 471.348, kemungkinan karena cuaca yang belum stabil, seperti hujan atau suhu dingin, yang membuat orang cenderung menghindari bersepeda. Pola ini menunjukkan bahwa faktor cuaca dan musim sangat mempengaruhi kebiasaan penyewaan sepeda.
